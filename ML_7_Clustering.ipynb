```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
x,y = make_blobs(n_samples = 1000, centers = 5, random_state = 42) # center = 5 means - generate 5 clusters
```

```python
pd.DataFrame(x)
```

```python
kmeans = KMeans(n_clusters = 5, random_state = 42)
kmeans.fit(x)
```

```python
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

```python
labels
```

```python
centroids
```

```python
plt.scatter(x[:,0], x[:,1], c = labels, cmap = 'viridis', s = 50, edgecolors = 'k')
plt.scatter(centroids[:,0], centroids[:,1], c = 'red', marker = 'x', s = 200, linewidth = 5, label = 'centroids')
plt.title('K-Means Clustering')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# distance between 2 centroids of clusters is called as inter centroid distance
# distance within the cluster is called as intracentroid
# less intracentroid distance is considered as better cluster
# bigger the centroid distance - then that is also considered good
```
